# Exploratory Data Analysis 

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
```

```python
train = pd.read_csv('datasets/train_cleaned.csv')
```

```python
train.head()
```

### Exploring min_experience, max_experience, and average_salary variables 

#### Variable distributions 

```python
plt.figure(figsize=(12,8))
plt.xlabel('min_experience in years')
plt.ylabel('frequency')
plt.hist(train['min_experience']);
plt.savefig('visuals/min_experience_dist.jpeg')
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('max_experience in years')
plt.ylabel('frequency')
plt.hist(train['max_experience']);
plt.savefig('visuals/max_experience_dist.jpeg')
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('average_salary')
plt.ylabel('frequency')
plt.hist(train['average_salary']);
plt.savefig('visuals/salary_dist.jpeg')
```

#### Scatterplots

```python
plt.figure(figsize=(12,8))
plt.xlabel('average_salary')
plt.ylabel('max_experience')
plt.scatter(y = train['average_salary'], x = train['max_experience'], alpha=0.6);
plt.savefig('visuals/salary_maxexperience_scatter.jpeg')
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('average_salary')
plt.ylabel('max_experience')
sns.stripplot(x="average_salary", y="max_experience", data=train, linewidth=1);
plt.savefig('visuals/salary_maxexperience_strip.jpeg')
```

```python
sns.swarmplot(x="average_salary", y="max_experience", data=train.sample(n=120), linewidth=1);
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('average_salary')
plt.ylabel('min_experience')
sns.stripplot(x="average_salary", y="min_experience", data=train, linewidth=1);
plt.savefig('visuals/salary_minexperience_strip.jpeg')
```

### key_skills EDA

```python
cv_skills = CountVectorizer()
```

```python
skills = train['key_skills']
```

```python
skills.head()
```

```python
X_skills = cv_skills.fit_transform(skills)
```

```python
skills_df = pd.DataFrame(X_skills.todense(), 
                          columns=cv_skills.get_feature_names_out())
```

```python
skills_df
```

#### Top 25 key_skills

```python
plt.figure(figsize=(12,8))
plt.xlabel('count')
plt.title('Top 25 skills')
skills_df.sum().sort_values(ascending=False).head(25).plot(kind='barh');

plt.savefig('visuals/top_25_skills.jpeg')
```

#### Explore skills with location 

```python
location_skills = skills_df
```

```python
location_skills['location'] = train['location']
```

```python
location_skills = location_skills.groupby('location').sum()
```

```python
location_skills.head()
```

```python
location_skills.shape
```
